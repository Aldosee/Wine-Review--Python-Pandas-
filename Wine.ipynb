### Answer Questions using Data Visualization

```python
# Question 
# 1. What is the relationship between wine points and price?
sns.lmplot(data = df_wine, x= 'price', y= 'points', hue = 'country')
```

```python
# 2. Which country has the most highly-rated wines (by points)?
# Group by 'country' and calculate the average points
avg_points_by_country = df_wine.groupby('country')['points'].mean()

# Plot the bar chart
avg_points_by_country.plot(kind='bar', figsize=(10, 6), color='skyblue')

# Add labels and title
plt.title('Average Wine Points by Country', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Average Points', fontsize=12)
```

```python
# 2.1 Showing the number of wines with points above 90 for each country:
# Filter the dataset for wines with points above 90
high_rated_wines = df_wine[df_wine['points'] > 90]

# Group by 'country' and count the number of wines
count_high_rated_by_country = high_rated_wines.groupby('country')['points'].count()

# Plot the bar chart
count_high_rated_by_country.plot(kind='bar', figsize=(10, 6), color='green')

# Add labels and title
plt.title('Number of Wines with Points Above 90 by Country', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Wines', fontsize=12)
```

```python
# 3. What are the most common wine varieties in the dataset?
# Group by 'variety' and count the number of occurrences
variety_counts_15 = df_wine['variety'].value_counts().head(15)

# Plot the bar chart
variety_counts_15.plot(kind='pie', figsize=(12, 10), autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Add labels and title
plt.title('Top 15 Wine Varieties Distribution', fontsize=16)
```

```python
# 4. Which wineries consistently produce the highest-rated wines?
# Calculate the average points by winery
avg_points_by_winery = df_wine.groupby('winery')['points'].mean()

# Sort and get the top 15 wineries
top_15_wineries = avg_points_by_winery.sort_values(ascending=False).head(15)

# Create the bar plot
ax = top_15_wineries.plot(kind='bar', figsize=(10, 6), color='orange')

# Set the title and labels
plt.title('Top 15 Wineries with the Highest Average Wine Points', fontsize=16)
plt.xlabel('Winery', fontsize=12)
plt.ylabel('Average Points', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add labels (average points) on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',  # Format the label with 2 decimal places
                (p.get_x() + p.get_width() / 2., p.get_height()),  # Position at the center-top of the bar
                ha='center', va='center', xytext=(0, 10),  # Offset the label by 10 points vertically
                textcoords='offset points')
```

```python
# 5. Is there a correlation between a wine's origin (country or province) and its price?
# A box plot to compare wine prices from different countries or provinces, showing the distribution and outliers.

# Set the figure size for the plot
plt.figure(figsize=(20, 20))

# Create a box plot using seaborn
sns.boxplot(x='country', y='price', data=df_wine)

# Add title and labels
plt.title('Wine Prices by Country', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Price', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')
```

```python
#6. Which countries or regions have the widest variety of wines?
# Count the number of wines for each variety in country
variety_by_country_top = df_wine.pivot_table(index='country', aggfunc='size', fill_value=0)

# Create the horizontal chart bar plot
variety_by_country_top.plot(kind='barh', stacked=True, figsize=(10, 12), colormap='tab20')

# Add labels and title
plt.title('Wine Varieties by Country (Horizontal)', fontsize=16)
plt.xlabel('Count of Wines', fontsize=12)
plt.ylabel('Country', fontsize=12)
```
